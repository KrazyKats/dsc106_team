```python
glucose_data = pd.read_json(Path(JSONS_DIR, 'flattened_glucose.json'))
glucose_data
```

```python
# Combine dataframes and get relevant timestamps
# 1. Get dataframe of timestamp intervals 1hr before meal to 1hr after meal and patient id
# To combine these dataframes, try using join asof
# 2. Combine this with glucose
# 3. Do aggregations and plot this data
def get_data_within_timeframe():
    merged_df = food_log_tagged.merge(glucose_data, left_on = 'ID', right_on = 'patient_id')
    return merged_df[(merged_df['datetime'] - merged_df['timestamp']).dt.total_seconds().abs() <= 3600]
```

```python
combined_df = get_data_within_timeframe()
```
