```python
def get_cleaned_dexcom():
    dexcom_3 = pd.read_csv('data/Dexcom_003.csv')
    # Remove duplicate column of index and remove initial missing rows
    dexcom_3 = dexcom_3.drop(columns = 'Index').iloc[12:]
    # Remove all columns with all na values
    dexcom_3 = dexcom_3[dexcom_3.columns[dexcom_3.isna().mean() < 1]]
    dexcom_3['Timestamp'] = pd.to_datetime(dexcom_3['Timestamp (YYYY-MM-DDThh:mm:ss)'])
    dexcom_3 = dexcom_3.drop(columns = ['Timestamp (YYYY-MM-DDThh:mm:ss)', 'Event Type'])
    return dexcom_3.set_index('Timestamp')
```

```python
import pandas as pd

# Assuming your dataframe is named 'df' with a DatetimeIndex and a column 'blood_sugar'
# If your dataframe is structured differently, adjust accordingly

def count_significant_variations(df, column='blood_sugar', threshold=30, window_minutes=30):
    """
    Count instances where max-min blood sugar within a 30-minute window is at least 30.
    
    Parameters:
    -----------
    df : pandas DataFrame
        DataFrame with DatetimeIndex and blood sugar readings
    column : str
        Name of the column containing blood sugar values
    threshold : int
        Minimum difference between max and min to count (default: 30)
    window_minutes : int
        Window size in minutes (default: 30)
    
    Returns:
    --------
    int : Count of windows with significant variations
    """
    # Ensure index is datetime
    if not isinstance(df.index, pd.DatetimeIndex):
        raise TypeError("DataFrame index must be DatetimeIndex")
    
    # Resample to 30-minute windows and compute min and max
    resampled = df.resample(f'{window_minutes}T')
    window_stats = pd.DataFrame({
        'max': resampled[column].max(),
        'min': resampled[column].min()
    })
    
    # Calculate difference between max and min
    window_stats['difference'] = window_stats['max'] - window_stats['min']
    
    # Count windows where difference exceeds threshold
    significant_count = (window_stats['difference'] >= threshold).sum()
    
    return significant_count

# Example usage with sample data (replace with your actual data)
# Create sample data

# Display sample of windows with their min/max values for verification
resampled = dexcom_3.resample('30min')
window_stats = pd.DataFrame({
    'max': resampled['Glucose Value (mg/dL)'].max(),
    'min': resampled['Glucose Value (mg/dL)'].min(),
    'difference': resampled['Glucose Value (mg/dL)'].max() - resampled['Glucose Value (mg/dL)'].min()
})
window_stats['significant'] = window_stats['difference'] >= 30

px.bar(window_stats.assign(date = window_stats.index.date).groupby('date')['significant'].sum(),
       title= 'Count of significant blood sugar spikes by day')
```
